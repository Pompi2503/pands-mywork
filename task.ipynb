```python
import numpy as np
import matplotlib.pyplot as plt
x = ('Jan', 'Feb', 'Mar', 'Apr')
y = (58000, 54000, 59000, 63000)


# Plot the graph
plt.plot(x, y, color='pink', marker='o', linestyle='-')
plt.yticks(np.arange(50000, 70000, 2000))

# Adjust the y-axis range to avoid exaggerated steep drops

plt.xlabel("Month")
plt.ylabel("Number of Toys Sold")
plt.title("Toys Sold Over Time")

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.show()
```
